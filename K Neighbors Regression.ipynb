This file has code for K Neighbors Regression and its accuracy will be compared with sklearn using mean squared error

```python
import pandas as pd
import numpy as np
from sklearn.datasets import load_boston
from sklearn.metrics import mean_squared_error

df = pd.read_csv("Advanced House Price Prediction.csv")
df.head()
```

```python
#Split train and test data
X_train = df.iloc[:int(len(df)*0.8), :-1]
Y_train = df.iloc[:int(len(df)*0.8), -1:]
X_test = df.iloc[int(len(df)*0.8):, :-1]
Y_test = df.iloc[int(len(df)*0.8):, -1:]
```

```python
#Creating KNeighborRegressor
class KNeighborRegressor:
    def __init__(self, n_neighbors):
        self.k=n_neighbors
        
    def fit(self,X,Y):
        self.X=X
        self.Y=Y
    
    def predict(self,test):
        y_pred = []
        distance = []
        for i in range(test.shape[0]):
            for j in range(0,self.X.shape[0]):
                #Calculating the euclidian distance between each new datapoint and all the existing data points and appending it to a list
                distance.append(np.linalg.norm(np.array(test.iloc[i]) - np.array(self.X.iloc[j])))
        distance_array = np.array(distance).reshape(-1,self.X.shape[0])
        for i in range(0,distance_array.shape[0]):
            #Finding K indices that are closer to the new datapoint (K=4 in this case)
            indices = np.sort(np.argpartition(distance_array[i], self.k)[:self.k])
            final_class = []
            for index in indices:
                #Find the output value of those K indices and calculate the mean value of it
                final_class.append(self.Y.loc[index, self.Y.columns][0])
            y_pred.append(np.mean(final_class))
        return np.array(y_pred)
            
model_KM = KNeighborRegressor(n_neighbors=4)
model_KM.fit(X_train,Y_train)
y_pred_km = model_KM.predict(X_test)

#Mean Squared Error to find the accuracy
from sklearn.metrics import mean_squared_error
print("Mean Sqaured Error: ",mean_squared_error(y_pred_km, Y_test))
```

```python
from sklearn.neighbors import KNeighborsRegressor
model_neigh = KNeighborsRegressor(n_neighbors=4)
model_neigh.fit(X_train, Y_train)
y_pred_neigh = model_KM.predict(X_test)

#Mean Squared Error to find the accuracy
print("Mean Sqaured Error using sklearn: ",mean_squared_error(y_pred_neigh, Y_test))
```

Both models have same mean squared error value which indicates that the first model is consistent with the sklearn model
